```python
import cv2
import os
```

```python

# Define the actions to be captured
actions = [
    "Smile", "Cheek_Puffing", "Open_Mouth", "Tongue_Out", "Brow_Raise", 
    "Eyes_Close", "Smile_Left", "Smile_Right", "Turn_Left", "Turn_Right", 
    "See_Down", "See_Up", "Head_Down_Left", "Head_Down_Right", 
    "Right_Eye_Close", "Left_Eye_Close", "Frown", "Show_Teeth", 
    "Pout", "Pout_Left", "Pout_Right", "Brow_Raise_with_Smile", 
    "Yawn", "Eye_Blink", "Inward_Lip_Press", "Surprise"
]
```

```python
# Define the base directory for saving images
base_dir = "facial_action_dataset"
os.makedirs(base_dir, exist_ok=True)

def capture_subject_images(subject_id):
    """Capture all actions for a single subject."""
    cap = cv2.VideoCapture(2)
    if not cap.isOpened():
        print("Error: Could not open webcam.")
        return

    print(f"Starting image capture for Subject {subject_id}")

    for action in actions:
        action_dir = os.path.join(base_dir, action)
        os.makedirs(action_dir, exist_ok=True)

        # Get the starting image number for this action
        existing_files = sorted(os.listdir(action_dir))
        start_number = 1
        if existing_files:
            start_number = int(existing_files[-1].split('.')[0]) + 1

        print(f"\nPrepare for action: {action}")
        print("Press 's' to capture an image, 'n' to skip this action, or 'q' to quit.")

        while True:
            ret, frame = cap.read()
            if not ret:
                print("Error: Could not read frame from webcam.")
                break

            # Display the live feed
            cv2.putText(frame, f"Action: {action} (Subject {subject_id})", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)
            cv2.putText(frame, "Press 's' to capture, 'n' to skip, 'q' to quit", (10, 70), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 1)
            cv2.imshow("Facial Action Capture", frame)

            key = cv2.waitKey(1) & 0xFF

            if key == ord('s'):  # Save an image
                file_path = os.path.join(action_dir, f"{start_number}.jpg")
                cv2.imwrite(file_path, frame)
                print(f"Saved: {file_path}")
                start_number += 1

                # Ask if another image should be taken for the same action
                print(f"Press 's' to capture another image for {action}, 'n' to move to the next action.")

            elif key == ord('n'):  # Skip to the next action
                print(f"Finished capturing for action: {action}")
                break

            elif key == ord('q'):  # Quit the entire process
                print("Exiting...")
                cap.release()
                cv2.destroyAllWindows()
                return

    print(f"Finished image capture for Subject {subject_id}.")
    cap.release()
    cv2.destroyAllWindows()
```

```python
# Main program loop
subject_id = input("Enter Subject ID: ")
capture_subject_images(subject_id)  
```
